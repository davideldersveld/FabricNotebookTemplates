### Visualizing Fabric AI Skill output in Notebook charts
Adapted from https://learn.microsoft.com/en-us/fabric/data-science/ai-skill-scenario by David Eldersveld

```python
import pandas as pd
import requests
import json
import pprint
from synapse.ml.mlflow import get_mlflow_env_config

def send_prompt(user_prompt, url):
    configs = get_mlflow_env_config()

    headers = {
        "Authorization": f"Bearer {configs.driver_aad_token}",
        "Content-Type": "application/json; charset=utf-8"
    }

    question = "{userQuestion: \"" + user_prompt + "\"}"

    response = requests.post(url, headers=headers, data = question)

    df = pd.json_normalize(response.json())

    df_result = pd.DataFrame(df["ResultRows"][0], columns=df["ResultHeaders"][0])
    return df_result
```

```python
user_prompt = "[Replace with prompt specific to your AI Skill]"
ai_skill_url = "[Replace with your AI Skill URL]"

df = send_prompt(user_prompt, ai_skill_url)

print(user_prompt)
display(df)
```
